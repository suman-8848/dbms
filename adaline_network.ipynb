```python

import numpy as np

input_values = np.array(
	[
		[-1, -1],
		[-1, 1],
		[1, -1],
		[1, 1]
	])
# output and gate
output_values = np.array([-1, -1, -1, 1])
print(input_values, output_values)
weight = np.array([0.3, 0.5])
bias = 0.3
learning_rate = 0.3
epoch = 100
for i in range(epoch):
  print("epoch :", i+1)
  sum_squared_error = 0.0

  for j in range(4):
    # yin = input_values[j][0]*weight[0] +input_values[j][1]*weight[1]+ bias
    yin = np.dot(input_values[j],np.transpose(weight))+bias
    # print(yin)
    error= output_values[j]-yin
    weight = weight + learning_rate*error*input_values[j]
    sum_squared_error+=error*error
    bias += learning_rate*error
    print("error: ",error)
  print("sum of squared error: ",sum_squared_error/4,"\n")
# validation
final_prediction = []
for j in range(4):
  # predict = input_values[j][0]*weight[0] +input_values[j][1]*weight[1]+ bias
  predict = np.dot(input_values[j],np.transpose(weight))+bias
  # if predict>=0:
  #   predict = 1
  # else:
  #   predict=0
  final_prediction.append(predict)
print("prediction: ",final_prediction)











```
